# mrm8488/distilroberta-finetuned-financial-news-sentiment-[link text](`***https://***`)analysis

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-classification", model="mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis")
```

```python
user_input = input("Enter a sentence to analyze sentiment: ")
```

```python
output = pipe(user_input)
print(output)
```

```python
# Display the output
print("Input Text:", user_input)
print("Output Label:", output[0]['label'])
print("Output Score:", output[0]['score'])
```
